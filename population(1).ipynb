```python
import pandas as pd
import matplotlib.pyplot as plt

#load data
df = pd.read_csv("population.csv")

print(df.head())
```

```python
#create a wide format with one column per city
pivot_df = df.pivot(index="Year", columns="City", values="Population")

#calculate the year-over-year percentage change in population for each city
pct_change_df = pivot_df.pct_change() * 100

print(pct_change_df)


```

```python
#create population growth rates over time for each city
pct_change_df.plot()
plt.xlabel("Year")
plt.ylabel("Population Growth Rate (%)")
plt.title("Annual Population Growth Rates for US Cities")
plt.show()
```

```python
#group the data by city and calculate the mean population for each city
city_populations = df.groupby("City")["Population"].mean()


print(city_populations)
```
