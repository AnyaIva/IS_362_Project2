```python
import pandas as pd
import matplotlib.pyplot as plt

#load data
df = pd.read_csv('Average_Hourly_Earnings.csv')

print(df.head())
```

```python
#calculate % increase from year to year
df_pct_yearly = ((df.iloc[-1] - df.iloc[0]) / df.iloc[0]) * 100

print("Percent increase from year to year:\n", df_pct_yearly)
```

```python
#calculate % increase from month to month
df_pct_monthly = ((df.iloc[-1] - df.iloc[0]) / df.iloc[0]) * 100 / 11

print("Percent increase from month to month:\n", df_pct_monthly)
```

```python
#plot the data

#set the 'Year' column as the index
df.set_index('Year', inplace=True)

#ceate a line plot of the hourly wages data
df.plot(kind='line', figsize=(12, 6))

#set plot title and axes labels
plt.title('Hourly Wages Over Time')
plt.xlabel('Year')
plt.ylabel('Hourly Wage ($)')


plt.show()
```
